### Finding missing values by the process Data Cleansing 

Here in the process of Data cleansing we will first find out the null values , then we will try to fix it by finding the mean and replacing it "true" value


```python
ebay.info() #with this function we can see all the info of dataset as we can see null count and its data type.
```

```python
ebay.isnull() #this function is helpful in finding the null vlues 
```

```python
ebay.isnull().sum() #this fuction help in finding the count of null values of the attribute which consist it.
```

```python
bidderrateof_mean = ebay.bidderrate.mean()
```

```python
print(bidderrateof_mean)
```

```python
ebay.bidderrate.fillna(bidderrateof_mean,inplace=True) #with this function we have replaced the null values to true.
```

```python
ebay.isnull().sum()#here after data cleaning we can see that we have rectified the "bidderrate" null value.
```

```python
# 1. HISTOGRAM - sns.histplot()
fig, ax = plt.subplots()
sns.histplot(x='price', data=ebay)
print("Minimum SP:",ebay['price'].min())
print("Maximum SP:",ebay['price'].max())
ax.set_xlim(10, 7000)
ax.set_ylim(0, 100)

plt.show()
```

```python
cols = 3
rows = 3
num_cols = ebay.select_dtypes(exclude='object').columns
fig = plt.figure( figsize=(cols*5, rows*5))
for i, col in enumerate(num_cols):
#for i in num_cols:    
    ax=fig.add_subplot(rows,cols,i+1)
    
    sns.histplot(x = ebay[col], ax = ax)
    
fig.tight_layout()  
plt.show()
```
